```python
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame (replace this with your actual DataFrame)
titanic = pd.DataFrame({
    'Class': ['First', 'Second', 'Third', 'First', 'Third', 'Second', 'Third', 'First']
})

# Count occurrences of each class
count = titanic['Class'].value_counts()

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Define the class names and colors
classes = ['First', 'Second', 'Third']
colors = ['yellow', 'blue', 'green']

# Plot the bar chart
# Ensure that all classes are accounted for, even if some have a count of 0
bar_counts = [count.get(c, 0) for c in classes]
ax.bar(classes, bar_counts, color=colors)

# Add labels and title (optional)
ax.set_xlabel('Class')
ax.set_ylabel('Count')
ax.set_title('Count of Each Class')

# Show the plot
plt.show()
```
