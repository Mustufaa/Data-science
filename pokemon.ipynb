```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df= pd.read_csv('Pokemon.csv')
df
```

```python
fig,ax=plt.subplots(figsize=(12,6))
#sns.barplot(x='Type 1', y='Name', data=df.head(10).reset_index(),ax=ax)
sns.countplot(data=df.head(10),x='Name',hue='Type 1')
plt.title('Top 10 Pokemon Types')
plt.xticks(rotation=90)
plt.show()
```

```python
#plt.figure(figsize=(10,7))
px.scatter_3d(data_frame=df,
              x='Type 1',
            y='Name',
            z='HP',
            color='Generation'
            )
#plt.title('3D graph on pokemon')
#plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data=df.head(20),
             x='HP',
             kde=True,
             bins=10
             )
plt.title('Histogram bases on pokemon HP')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.swarmplot( 
    data=df,
    x='Generation',
    y='HP',
    hue='Type 2',
    dodge=True
)
plt.title('Swarm Plot of HP by Generation and Type 2')
plt.show()
```

```python

px.sunburst(
    title='pokemon plot of HP,Attack by Generation ',
    data_frame=df.head(5),
    path=['Name','HP','Attack','Speed'],
    values='Generation',
)
```

```python
Name=df['Name'].head(10)
HP=df['HP'].head(10)
plt.pie(HP,labels=Name,autopct='%.1f%%')
plt.title('Pie Chart of First 10 Rows')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df.head(100),x='Type 1',y='HP')
plt.title('Boxplot of HP by Type 1')
plt.show()
```

```python
fire=df.groupby(by='Type 1').get_group('Fire')
fire
```

```python
fire=df.groupby(by='Type 1').get_group('Fire')
fire
plt.bar(fire['Generation'], fire['HP'])
plt.xlabel('Generation')
plt.ylabel('HP')
plt.title('Histogram of fire pokemon')

plt.show()
```

```python
sns.scatterplot(x=fire['Attack'],y=fire['Defense'])
```

```python
fig,ax=plt.subplots(figsize=(15,10))
sns.lineplot(data=fire['Speed'],dashes=False)
```
